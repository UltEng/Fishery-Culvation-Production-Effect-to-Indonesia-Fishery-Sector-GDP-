# Luas Lahan Budidaya

```python
df_luas_raw =pd.read_excel('drive/MyDrive/Capstone Tetris/Luas Lahan.xlsx')
```

```python
df_luas_raw.head()
```

```python
df_luas_raw.shape
```

```python
luas = df_luas_raw.transpose().iloc[1].to_numpy()
i = 2
while i < df_luas_raw.shape[1]:
    data= df_luas_raw.transpose().iloc[i].to_numpy()
    luas = np.concatenate((luas, data))
    i = i + 1
print(luas)
print(len(luas))
```

```python
from itertools import repeat
year = []
i = 2011
while i < 2021:
    year.extend(repeat(i,df_luas_raw.shape[0]))
    i = i + 1
print(year)
print(len(year))
```

```python
provinsi = df_luas_raw.transpose().iloc[0].to_numpy()
i = 2
while i < df_luas_raw.shape[1]:
    data = df_luas_raw.transpose().iloc[0].to_numpy()
    provinsi = np.concatenate((provinsi, data))
    i = i + 1
print(provinsi)
print(len(provinsi))
```

```python
dict1 = {'provinsi' : provinsi, 'tahun' : year, 'luas lahan' : luas }
luas = pd.DataFrame(dict1, columns=['provinsi','tahun','luas lahan'])
luas = luas.drop(luas.index[0])
luas.head()
```

# Rumah Tangga Perikanan

```python
df_rtpb_raw =pd.read_excel('drive/MyDrive/Capstone Tetris/RTP Budidaya.xlsx')
df_rtpl_raw =pd.read_excel('drive/MyDrive/Capstone Tetris/RTP Laut.xlsx')
df_rtpp_raw =pd.read_excel('drive/MyDrive/Capstone Tetris/RTP PUD.xlsx')
```

```python
df_rtpb_raw.head()
```

```python
df_rtpb_raw.shape
```

```python
jum = []
i = 1
j = 1
k = 1
while i < df_rtpb_raw.shape[1]:
    data= df_rtpb_raw.transpose().iloc[i].to_numpy()
    jum = np.concatenate((jum, data))
    i = i + 1
while j < df_rtpb_raw.shape[1]:
    data= df_rtpl_raw.transpose().iloc[j].to_numpy()
    jum = np.concatenate((jum, data))
    j = j + 1
while k < df_rtpb_raw.shape[1]:
    data= df_rtpp_raw.transpose().iloc[k].to_numpy()
    jum = np.concatenate((jum, data))
    k = k + 1
print(jum)
print(len(jum))
```

```python
tahun = []
i = 2011
j = 2011
k = 2011
while i < 2021:
    tahun.extend(repeat(i,df_rtpb_raw.shape[0]))
    i = i + 1
while j < 2021:
    tahun.extend(repeat(j,df_rtpb_raw.shape[0]))
    j = j + 1
while k < 2021:
    tahun.extend(repeat(k,df_rtpb_raw.shape[0]))
    k = k + 1
print(tahun)
print(len(tahun))
```

```python
prov = []
i = 1
j = 1
k = 1
while i < df_rtpb_raw.shape[1]:
    data = df_rtpb_raw.transpose().iloc[0].to_numpy()
    prov = np.concatenate((prov, data))
    i = i + 1
while j < df_rtpb_raw.shape[1]:
    data = df_rtpb_raw.transpose().iloc[0].to_numpy()
    prov = np.concatenate((prov, data))
    j = j + 1
while k < df_rtpb_raw.shape[1]:
    data = df_rtpb_raw.transpose().iloc[0].to_numpy()
    prov = np.concatenate((prov, data))
    k = k + 1
print(prov)
print(len(prov))
```

```python
jenis_rtp = []
jenis_rtp.extend(repeat('Budidaya', 340))
jenis_rtp.extend(repeat('Laut', 340))
jenis_rtp.extend(repeat('PUD', 340))
print(jenis_rtp)
print(len(jenis_rtp))
```

```python
dict2 = {'provinsi' : prov, 'tahun' : tahun,'jenis' : jenis_rtp, 'jumlah RTP' : jum }
rtp = pd.DataFrame(dict2, columns=['provinsi','tahun','jenis', 'jumlah RTP'])
rtp = rtp.drop(rtp.index[0])
rtp.head()
```

# Korelasi Variabel

```python
prod_corr = prod_group[prod_group['Jenis Usaha'] == 'BUDIDAYA']
prod_corr.drop('Jenis Usaha', axis=1, inplace=True)
prod_corr.head()
```

```python
jum_corr = jum_group[jum_group['Jenis Usaha'] == 'BUDIDAYA']
jum_corr.drop('Jenis Usaha', axis=1, inplace=True)
jum_corr.head()
```

```python
luas_corr = luas.groupby(['tahun']).sum()['luas lahan']
luas_corr.to_frame()
luas_corr = luas_corr.reset_index(level=['tahun'])
luas_corr = luas_corr.rename({'tahun':'Tahun', 'luas lahan':'Luas Lahan'}, axis =1)
luas_corr.head()
```

```python
rtp_corr = rtp_group[rtp_group['jenis'] == 'Budidaya']
rtp_corr.drop('jenis', axis=1, inplace=True)
rtp_corr = rtp_corr.rename({'tahun':'Tahun','jumlah RTP':'Jumlah RTP'}, axis =1)
rtp_corr.head()
```

```python
corr = prod_corr.merge(jum_corr, on='Tahun').merge(luas_corr, on='Tahun').merge(rtp_corr, on='Tahun')
corr.head()
```

```python
import plotly.figure_factory as ff
df_corr = corr[['Nilai Produksi', 'Volume Produksi', 'Luas Lahan', 'Jumlah RTP']].corr()
x = list(df_corr.columns)
y = list(df_corr.index)
z = np.array(df_corr)

fig = px.imshow(z, x=x, y=y, color_continuous_scale='Viridis', aspect="auto")
fig.update_traces(text=np.around(z, decimals=2), texttemplate="%{text}")
fig.update_xaxes(side="top")
fig.show()
```

```python
corr1 = prod_corr.merge(pdb_group, on='Tahun')
corr1.head()
```

```python
fig = px.scatter(corr1, x="Nilai Produksi", y="PDB")
fig.show()
```

```python
from scipy.stats import pearsonr

corr, _ = pearsonr(corr1["Nilai Produksi"], corr1["PDB"])
print('Pearsons correlation: %.3f' % corr)
if corr == 0:
  print('Tidak ada hubungan sama sekali')
elif corr > 0 and corr < 0.2:
  print('Hubungan sangat lemah')
elif corr >= 0.2 and corr < 0.4:
  print('Hubungan lemah')
elif corr >= 0.4 and corr < 0.6:
  print('Hubungan cukup kuat')
elif corr >= 0.6 and corr < 0.8:
  print('Hubungan kuat')
elif corr >= 0.8 and corr < 1:
  print('Hubungan sangat kuat')
elif corr == 1:
  print('Hubungan sempurna')
```
